```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

df = pd.read_csv('data_clean.csv')
df
```

```python
df.columns
```

```python
plt.scatter(df['ALLSKY_SFC_UV_INDEX'], np.log(df['TS']), alpha = 0.1)
plt.xlabel('UV_INDEX', fontsize = 20)
plt.ylabel('TS', fontsize = 20)
plt.show()
```

```python
plt.scatter(df['ALLSKY_SFC_UVA'], np.log(df['TS']), alpha = 0.1)
plt.xlabel('UVA', fontsize = 20)
plt.ylabel('TS', fontsize = 20)
plt.show()
```

```python
plt.scatter(df['ALLSKY_SFC_UVB'], np.log(df['TS']), alpha = 0.1)
plt.xlabel('UVB', fontsize = 20)
plt.ylabel('TS', fontsize = 20)
plt.show()
```

```python
plt.plot(df['DATE'].astype(str), df['TS'])
plt.xlabel('DATE', fontsize = 20)
plt.ylabel('TS', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
periode_smoothing = 30
df['Smoothed_TS'] = df['TS'].rolling(periode_smoothing).mean()

plt.scatter(df['DATE'].astype(str), df['Smoothed_TS'], alpha = 0.1)
plt.xlabel('DATE', fontsize = 20)
plt.ylabel('TS', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.scatter(df['DATE'].astype(str), df['TS'], alpha = 0.1)
plt.xlabel('DATE', fontsize = 20)
plt.ylabel('TS', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```
